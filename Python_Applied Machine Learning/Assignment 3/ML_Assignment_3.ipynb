# **SeoulBikeData (Dataset_1)**

### **Data Processing**

```
dataset = pd.read_csv("/content/SeoulBikeData.csv",encoding='latin1') 
dataset.head(3) # view the data
# dataset.dtypes
```

```
X = dataset.iloc[:, 2:-1].values #Drop data
y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(y)
```

Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1,-2])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X[0,:]) # 15 Variables
```

Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
import collections
collections.Counter(y)
```

Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```
print(y_train.shape, X_train.shape, y_test.shape, X_test.shape)
```

Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# ready X_train, X_test, y_train, y_test
```

##**Artificial neural networks (ANN)**

### **Building the Ann with 2 hidden layers 4 & 6 Units 100 & 300 epochs**

 **Initializing the ANN
(2 hidden layer, 6 unite, activation='relu', output activation='sigmoid')**

```
ann_2_6 = tf.keras.models.Sequential()
ann_2_6.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann_2_6.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann_2_6.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

(2 hidden layer, 4 unite, activation='relu', output activation='sigmoid')

```
ann_2_4 = tf.keras.models.Sequential()
ann_2_4.add(tf.keras.layers.Dense(units=4, activation='relu'))
ann_2_4.add(tf.keras.layers.Dense(units=4, activation='relu'))
ann_2_4.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

#### **ANN 2 hidden layers 6 units "Adam" epochs 100**

Compiling the ANN

```
ann_2_6.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Training the Training set

```
ann_2_6.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

Training set results and Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred_in = ann_2_6.predict(X_train)
y_pred_in = (y_pred_in > 0.5)
cm = confusion_matrix(y_train, y_pred_in)
print(cm)
print("Accuracy: {:.2f} %".format(accuracy_score(y_train, y_pred_in)*100))
```

Test set Predition & Confusion Matrix

```
y_pred = ann_2_6.predict(X_test)
y_pred = (y_pred > 0.5)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print("Accuracy: {:.2f} %".format(accuracy_score(y_test, y_pred)*100))
```
